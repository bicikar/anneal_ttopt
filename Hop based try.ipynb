В текущем ноутбуке представлены результаты работы квантовых подходов edge- и hop-based, и алгоритма TTopt, а также их сравнение результатов данных подходов.

Работаем с сетчатыми графами шириной $w$ и высотой $h$

Вершин в графе: $w \cdot h$

Ребер в графе: $(w - 1)\cdot h + w\cdot (h - 1) + 2(w - 1)(h-1)$

$\mathbf{P}$ ~ $w + h$

Hop-based qubits: $\mathbf{P}\mathbf{V} = w^2\cdot h + w\cdot h^2$

Edge-based qubtis: $\mathbf{V} + \mathbf{E} = wh + (w - 1)\cdot h + w\cdot (h - 1) + 2(w - 1)(h-1)$

# Простой граф

Для начала используем простой граф на рисунке:

```python
def simple_graph():
    graph = Graph(4)
    graph.add_edge(0, 1, 5)
    graph.add_edge(0, 2, 5)
    graph.add_edge(1, 2, 2)
    graph.add_edge(1, 3, 2)
    graph.add_edge(2, 3, 10)
    
    return graph
graph = simple_graph()
```

![title](data/graph_example.jpg)

```python
field = Field(2, 2, start=(0,0), end=(1, 1), graph=simple_graph())
```

Число переменных у edge-based подхода:

```python
len(field.edge_based.Q)
```

Число переменных у hop-based подхода:

```python
len(field.hop_based.Q)
```

Используем $3\cdot10^5$ обращений к функции для TTopt, каждый подход прогоняется 100 раз

```python
correct_path, admissible, identical, costs, true_shortest = run_stats(field, 3e+5, 40000, 50000)
```

```python
draw_table_stat(correct_path, true_shortest, admissible, identical)
```

```python
draw_big_boxplot(costs)
```
